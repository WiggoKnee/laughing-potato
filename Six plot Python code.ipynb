```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the excel file
df = pd.read_excel("Final CTEC 128 Stats.xlsx")

# Clean and prepare data
df.columns = df.columns.str.strip()
df.rename(columns={'Unnamed: 0': 'Crime'}, inplace=True)

races = ['Black', 'White', 'Pacific Islander', 'American Indian or Alaska Native', 'Asian']
plt.figure(figsize=(10, 6))
plt.stackplot(df['Crime'], [df[race] for race in races], labels=races)
plt.title('Stack Plot: Crime Types')
plt.xlabel('Crime')
plt.ylabel('Number of Offenders')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the excel file
df = pd.read_excel("Final CTEC 128 Stats.xlsx")

#Dataset
df.columns = df.columns.str.strip()
df.rename(columns={'Unnamed: 0': 'Crime'}, inplace=True)

races = ['Black', 'White', 'Pacific Islander', 'American Indian or Alaska Native', 'Asian']

#Chart reqs
fig, axs = plt.subplots(len(races), 1, figsize=(10, 12), sharex=True)
for i, race in enumerate(races):
    axs[i].bar(df['Crime'], df[race], color='skyblue')
    axs[i].set_title(f'{race} Offenders')
    axs[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

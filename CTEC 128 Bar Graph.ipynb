```python
# Aaronea Wiggins Histogram plot - CTEC 298 | 
import matplotlib.pyplot as plt

# Details of Dataset
categories = ['Black', 'White', 'Pacific Islander', 'American Indian', 'Asian']
values = [728044, 1676290, 2404334, 48587, 30845]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='green', edgecolor='black')

# X-axis and Y-axis titles and Title of overall Histogram plot
plt.xlabel('Race')
plt.ylabel('Amount of crimes per race')
plt.title('2020 Juvenile Crimes')

# Show plot results
plt.tight_layout()
plt.show()
```
